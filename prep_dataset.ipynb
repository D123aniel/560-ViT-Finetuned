```python
# Shuffles data in a directory, but you have to do this for every directory
# Does not go into subdirectories

import os
import random

def shuffle_in_dir(dir_path):
    dir = os.listdir(dir_path)
    random.shuffle(dir)
    return dir

file_dir = "../dataset/Real faces"
shuffled_files = shuffle_in_dir(file_dir)

for file in shuffled_files:
    file_path = os.path.join(file_dir, file)

print("Shuffled", file_dir, "without errors")
```

```python
from datasets import load_dataset, DatasetDict

dataset = load_dataset("imagefolder", data_dir="../dataset")['train']

# Split dataset into 6k training, 2k validation, 2k test

train_remainder_split = dataset.train_test_split(test_size=4000, stratify_by_column="label", seed=42)

val_test_split = train_remainder_split["test"].train_test_split(test_size=2000, stratify_by_column="label", seed=42)

final_dataset = DatasetDict({
    "train": train_remainder_split["train"],
    "validation": val_test_split["train"],
    "test": val_test_split["test"]
})

print(final_dataset)
from collections import Counter
print(f"Train: {Counter(final_dataset['train']['label'])}")
```

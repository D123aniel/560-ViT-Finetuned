```python
# Tester

from transformers import pipeline
from PIL import Image
import base64
import requests
from io import BytesIO

classifier = pipeline("image-classification", model="D123aniel/560_ViT_Finetuned")

data_url = "https://cdn.discordapp.com/attachments/711030003420364833/1495637394584637470/image.png?ex=69e6f85e&is=69e5a6de&hm=966fa0ebd20c5546666e8ead54db18cea3e6d96b2236a18e1fa996e85fc9bbea&"
response = requests.get(data_url)
if response.status_code == 200:
    # Convert the bytes from the web request into a PIL Image
    image = Image.open(BytesIO(response.content))
    
    # 3. Run Inference
    results = classifier(image)
    
    print("\n--- Classification Results ---")  
    for result in results:
        # result is a dict like {'label': 'REAL', 'score': 0.99}
        print(f"Label: {result['label']} | Confidence: {result['score']:.4f}")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```
